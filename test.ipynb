```python
from QHyper.problems import KnapsackProblem
from QHyper.solvers import VQA, solver_from_config

problem = KnapsackProblem(max_weight=2, item_weights=[1, 1, 1], item_values=[2, 2, 1])

problem_config = {
    'type': 'knapsack',
    'max_weight': 2,
    'item_weights': [1, 1, 1],
    'item_values': [2, 2, 1],
}


params_config = {
    'angles': [[0.5]*5, [1]*5],
    'hyper_args': [1, 2.5, 2.5],
}

solver_config = {
    'optimizer': {
        'type': 'scipy',
        'maxfun': 200,
        "bounds": [(0, 2 * 3.1415)] * 10
    },
    'pqc': {
        'type': 'wfqaoa',
        'layers': 5,
    },
    'hyper_optimizer': {
        'type': 'random',
        'processes': 4,
        'number_of_samples': 100,
        'bounds': [(1, 10), (1, 10), (1, 10)]
    },
    'params_inits': params_config
}

full_solver_config = {
    "solver": {
        "type": "vqa",
        **solver_config
    },
    "problem": problem_config
}

vqa = solver_from_config(full_solver_config)
vqa = VQA.from_config(problem, solver_config)
```

```python
solver_results = vqa.solve()

print()
```
